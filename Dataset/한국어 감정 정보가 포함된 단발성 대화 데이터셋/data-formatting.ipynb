### 한국어 감정 정보가 포함된 단발성 대화 데이터셋

```python
# 데이터 불러오기

df_short = pd.read_csv('./한국어_단발성_대화_데이터셋.csv')
```

```python
# 감정 데이터 숫자로 변환하기

emotion_dict = {'행복': 0, '분노': 1, '혐오': 2, '공포': 3, '중립': 4, '슬픔': 5, '놀람': 6}
emotion_list = ['행복', '분노', '혐오', '공포', '중립', '슬픔', '놀람']
emotions = [] # [행복, 분노, 혐오, 공포, 중립, 슬픔, 놀람]

for i in range(len(df_short)):
    df_short.loc[i, 1] = emotion_dict[df_short.iloc[i, 1]]
```
