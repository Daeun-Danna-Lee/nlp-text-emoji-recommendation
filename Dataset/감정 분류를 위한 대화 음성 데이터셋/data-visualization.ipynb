```python
import pandas as pd
from pandas import DataFrame as df
import csv
```

```python
# 파일 불러오기

df_4   = pd.read_csv('./4차년도.csv',encoding='cp949')
df_5   = pd.read_csv('./5차년도.csv',encoding='cp949')
df_5_2 = pd.read_csv('./5차년도_2차.csv',encoding='cp949')
```

```python
# 모든 연도 데이터 합치기

merged_df = pd.concat([df_4, df_5, df_5_2])
merged_df = merged_df.reset_index()
```

```python
# 필요 없는 column 삭제하기

data = merged_df.drop(columns=['index', 'wav_id', '나이', '성별', '상황'], axis=0)
```

```python
data
```

```python
# 통합된 하나의 감정 추출하기

emotion_dict = {'happy': 0, 'angry': 1, 'disgust': 2, 'fear': 3, 'neutral': 4, 'sad': 5, 'surprise': 6}
emotion_list = ['happiness', 'angry', 'disgust', 'fear', 'neutral', 'sadness', 'surprise']
emotions = [] # [happy, angry, disgust, fear, neutral, sad, surprise]

for i in range(len(data)):
    emotions = [0 for _ in range(7)]
    
    for j in range(1, 10, 2):
        emotions[emotion_list.index(data.iloc[i, j].lower())] += 1
    
    data.loc[i, '평균감정'] = emotion_list[emotions.index(max(emotions))]
```

```python
data
```

```python
# 필요 없는 column 삭제하기

data = data.drop(columns=['1번 감정', '1번 감정세기', '2번 감정', '2번 감정세기', '3번 감정', '3번 감정세기', '4번 감정', '4번감정세기', '5번 감정', '5번 감정세기'], axis=0)
data
```

```python
# 데이터 저장

path = './processed_data'
data.to_csv(path , sep=',' , header=None , index=None)
```
